So, now we have a dictionary. We need to extract the information inside "g" (gender). Let's do that. 

```python
gender = inspecting_characters["g"]
gender
```

So, in there, inference means the probabily of the gender inferred for the character, and **argmax** (what we are looking for) is the inferred gender of the character. so what we need to do is to extract argmax and count how many times we have male ("he/him/his") or female ("she/her") in our book.

```python
type(gender)
```

```python
arg_max = gender["argmax"]
arg_max
```

That being said, we also need to inspect the id dictionary to make sure that we are counting the gender of each individual character. 

```python
character_id = inspecting_characters["id"]
character_id
```

So this means that each element in the "characters" dictionary refers to 1 character (and therefore there is coreference resolution). So it is safe to assume that if we count the "argmax" number of mentions male vs. female we are getting a classification on the referencial gender of unique characters (so "Tom Sawyer", "He", "The Kid" are all the same id:26 and therefore all those mentions are including in "argmax" : "he/him/his".

Now, let's scale all that to extract just the "argmax" part of our file and count how many times it is male and how many times it is female.

```python
characters = book_data["characters"]

gender_counts = {}

for ch in characters:
    g = ch.get("g") or {}
    gender = g.get("argmax")
    if gender:
        gender_counts[gender] = gender_counts.get(gender, 0) + 1

gender_counts
```
