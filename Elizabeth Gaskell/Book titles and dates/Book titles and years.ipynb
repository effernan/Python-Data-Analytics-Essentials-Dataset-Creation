In this notebook we are going to put together all the book titles and dates together!

```python
import pandas as pd
```

```python
book_names = ["A Dark Night's Work", "Cranford", "Mary Barton", "My Lady Ludlow", "North and South", "Ruth",
              "Sylvia's Lovers", "Wives and Daughters"]
```

```python
dates = [1863, 1853, 1848, 1859, 1855, 1853, 1863, 1866]
```

```python
len(book_names)
```

```python
len(dates)
```

```python
data = pd.DataFrame({"Titles" : book_names, "Year" : dates})
```
