In this notebook we are going to put together all the book titles and dates together!

```python
import pandas as pd
```

```python
book_names = ["A Son at the Front", "Summer", "The Age of Innocence", "The Children", "The Custom of the Country", 
              "The Fruit of the Tree", "The Glimpses of the Moon", "The House of Mirth", "The Mother's Recompense",
              "The Reef", "The Valley of Decision", "Twilight Sleep"]
```

```python
dates = [1923, 1917, 1920, 1928, 1913, 1907, 1922, 1905, 1925, 1912, 1902, 1927]
```

```python
len(book_names)
```

```python
len(dates)
```

```python
data = pd.DataFrame({"Titles" : book_names, "Year" : dates})
```
